## 💼 Problema de Negócio

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/superstore.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df_region = df.groupby('Region')['Sales'].sum().sort_values()
```

```python
plt.figure(figsize=(8,5))
df_region.plot(kind='barh')
plt.title('Revenue by Region')
plt.xlabel('Sales')
plt.ylabel('Region')
plt.savefig('revenue_by_region.png')
plt.show()
```

A análise mostra como a receita está distribuída entre as regiões.

Observa-se que algumas regiões concentram maior volume de vendas, indicando maior presença de mercado.

```python
df['Order.Date'] = pd.to_datetime(df['Order.Date'])

df_time = df.groupby('Order.Date')['Sales'].sum()

plt.figure(figsize=(10,5))
df_time.plot()
plt.title('Sales Over Time')
plt.xlabel('Order Date')
plt.ylabel('Sales')
plt.savefig('sales_over_time.png')
plt.show()
```

O gráfico permite visualizar a evolução das vendas ao longo do tempo, ajudando a identificar tendências.

```python
df_category = df.groupby('Category')['Sales'].sum()

df_category.plot(kind='bar')
plt.title('Sales by Category')
plt.savefig('sales_by_category.png')
plt.show()
```

As categorias apresentam diferenças no volume de vendas, indicando quais áreas possuem maior participação.

```python
df_profit = df.groupby('Region')['Profit'].sum()

df_profit.plot(kind='bar')
plt.title('Profit by Region')
plt.savefig('profit_by_region.png')
plt.show()
```

O lucro por região mostra não apenas volume de vendas, mas eficiência financeira.

## 📌 Conclusão

- A análise mostrou diferenças de desempenho entre regiões
- Algumas categorias apresentam maior volume de vendas
- O comportamento das vendas ao longo do tempo permite identificar padrões
